6) Utilizar un ciclo while para realizar el factorial de un número guardado en una variable, sólo si la variable contiene un número entero mayor a 0

```python
n = 5
if (type(n) == int):#Verifica que n sea un número entero. Si le pasas un texto como "5", saltará directamente al último else y dirá que no es un entero.
    if (n > 0):#na vez confirmado que es entero, revisa que sea positivo. Si pones un número negativo, te avisará que no es mayor a cero.
        factorial = n# Se crea la variable factorial y se le asigna el valor inicial de n (o sea, 5).
        while (n > 2):#El ciclo se ejecutará mientras n sea mayor a 2.¿Por qué hasta 2? Porque multiplicar por 1 no cambia el resultado, así que el código ahorra una vuelta de bucle por eficiencia.
            n = n - 1 #Al usar n = n - 1 antes de la multiplicación, el código tiene que estar muy bien coordinado para no saltarse el primer número o multiplicar por un número de más. En este caso, funciona perfectamente porque inicializaste factorial = n afuera del bucle.
            factorial = factorial * n
        print('El factorial es', factorial)
    else:
        print('La variable no es mayor a cero')
else:
    print('La variable no es un entero')
```
