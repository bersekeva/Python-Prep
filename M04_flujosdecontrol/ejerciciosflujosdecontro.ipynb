```python
x = 5
while True:
    x -= 1
    print (x)
    if x == 0:
        break
    print("Fin del bucle")
#4
#Fin del bucle
#3
#Fin del bucle
#2
##Fin del bucle
#1
#Fin del bucle
#0
#__________________
#El problema está en la indentación (los espacios al inicio de la línea). En Python, la indentación le dice al lenguaje qué instrucciones están "dentro" del bucle y cuáles están fuera.
#En tu código, la línea print("Fin del bucle") tiene espacios a la izquierda, lo que le indica a Python que esa instrucción es parte de las tareas que debe repetir en cada vuelta del while.
#La solución:
#Para que ese mensaje solo aparezca una vez al final, debes quitarle los espacios a la izquierda. Al alinearlo con la palabra while, Python entiende que debe ejecutarlo solo cuando el bucle termine por completo.
```
