1) A partir de una lista vacía, utilizar un ciclo while para cargar allí números negativos del -15 al -1

```python
lista1= []
n = -15
while n < 0:
    lista1.append(n)
    n += 1
```

```python
print(lista1)
```

2) ¿Con un ciclo while sería posible recorrer la lista para imprimir sólo los números pares?

```python
n = 0  # Inicializa la variable n con el valor 0

while n < len(lista1):  # Mientras n sea menor que la longitud de la lista, se ejecuta el bucle
    if lista1[n] % 2 == 0:  # Verifica si el elemento en la posición n de la lista es divisible por 2
        print(lista1[n])  # Imprime el elemento si cumple la condición
    n += 1  # Incrementa el valor de n en 1 en cada iteración
```

3) Resolver el punto anterior sin utilizar un ciclo while

```python
for par in lista1:
    if par % 2 == 0:
        print(par)
```

5) Utilizar la función **enumerate** para obtener dentro del iterable, tambien el índice al que corresponde el elemento

7) La sucesión de Fibonacci es un listado de números que sigue la fórmula: <br>
n<sub>0</sub> = 0<br>
n<sub>1</sub> = 1<br>
n<sub>i</sub> = n<sub>i-1</sub> + n<sub>i-2</sub><br>
Crear una lista con los primeros treinta números de la sucesión.<br>

```python
fibo = [0,1]  #Creas la lista con los dos números obligatorios. El tamaño inicial de esta lista es 2.
n = 2 # Esta es tu variable de control. ¿Por qué empieza en 2? Porque en Python los índices empiezan en 0. Como ya tienes ocupadas las posiciones 0 y 1, el próximo número que vas a calcular debe guardarse en la posición 2.

while n <=30:
    fibo.append( fibo[n-1] + fibo[n-2])
#¿Qué hace Python entonces?
# Busca qué hay en fibo[1], Hay un 1.
# Busca qué hay en fibo[0] Hay un 0.
#Los suma: $1 + 0 =1
#Aplica el .append(1), lo que mete ese resultado al final de la lista.La lista ahora queda así: [0, 1, 1] (el último 1 está en la posición 2).
    n+=1
#El contador sube. Ahora n vale 3.
print(fibo)
```

9) La proporción aurea se expresa con una proporción matemática que nace el número irracional Phi= 1,618… que los griegos llamaron número áureo. El cuál se puede aproximar con la sucesión de Fibonacci. Con la lista del ejercicio anterior, imprimir el cociente de los últimos 5 pares de dos números contiguos:<br>
Donde i es la cantidad total de elementos<br>
n<sub>i-1</sub> / n<sub>i</sub><br>
n<sub>i-2</sub> / n<sub>i-1</sub><br>
n<sub>i-3</sub> / n<sub>i-2</sub><br>
n<sub>i-4</sub> / n<sub>i-3</sub><br>
n<sub>i-5</sub> / n<sub>i-4</sub><br>
 

```python
i = len(fibo)  # i vale 30
control = 1    # Nuestro contador para el bucle

while control <= 5:
    # 1. Hacemos la división usando 'control' para retroceder posiciones
    numerador = fibo[i - 1 - control]
    denominador = fibo[i - control]
    print(numerador / denominador)
    
    # 2. El paso de avance para no quedar atrapados en un bucle infinito
    control += 1
```
