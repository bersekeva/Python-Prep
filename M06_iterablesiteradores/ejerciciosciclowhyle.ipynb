1.Realizar la suma de numero multiplos de 5 menores de 50.
1.2 Del ejercicio anterior, solicita el número limite al celular

```python
variabledeinicio = 0
suma = 0
while variabledeinicio <= 50:
    suma = suma + variabledeinicio # el resultado en el primer ciclo es 0
    variabledeinicio += 5 # el resultado en el segundo ciclo en 5

print ('la suma es: ', suma)

#Es completamente normal que te cueste racionalizar el orden del while. A diferencia del ciclo for (que es más automático), el while es como un mecanismo de relojería manual: si mueves una pieza de lugar, todo el resultado cambia.
#El orden del print o de la operación matemática depende de una sola pregunta: ¿Quieres capturar el valor antes o después de que cambie?.
#-Osea si el print va antes, te va a entregar el resultado que da cada ciclo, no el resultado final del ciclo.
```

2.Mostrar la tabla de multiplicar del número 7
Ej: 7 * 1 = 7
    7 * 2 = 14

```python
tabla = 7
inicio = 1

while inicio <= 10:
    resultado = tabla * inicio
    print(tabla, '*', inicio, '=', resultado)
    inicio += 1
```
