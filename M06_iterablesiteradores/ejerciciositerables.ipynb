```python
#Este código es un ejemplo clásico de cómo recorrer una lista utilizando un ciclo while en lugar de un for. Es una simulación manual de cómo Python lee los elementos uno por uno usando sus índices (las posiciones numéricas).

#Aquí tienes la explicación de su estructura y el paso a paso detallado:

lista1=[1,3,4,6,7]#Contiene los datos. Recuerda que en Python las posiciones empiezan a contarse desde el 0. Así que el 1 está en la posición 0, el 3 en la 1, el 4 en la 2, etc.
i=0#Es la variable que usaremos como "puntero" o "índice". Empieza en 0 porque queremos arrancar desde el primer elemento.
while i<len(lista1):#calcula el tamaño de la lista, que en este caso es 5 (porque tiene 5 elementos). La condición dice: "Sigue ejecutando este bloque mientras i sea menor que 5". Esto evita que intentemos buscar una posición que no existe.
    elemento = lista1[i]
    print(elemento)#
    i += 1#
```

```python
lista2 = [1,3,4,6,7]
for elemento in lista2:
    print(elemento)
#Es lo mismo que el ejercicio anterior, pero mucho mas simplificado.
```

```python
from collections.abc import Iterable
#Aquí estás importando una herramienta del sistema de Python llamada Iterable. Un "iterable" es cualquier objeto en Python que se puede recorrer elemento por elemento dentro de un bucle (como un ciclo for).
cadena = 'Henry'
#Creas una variable de tipo texto (str).
numero = 9
#Creas una variable de tipo número entero (int).
#_____________________________________________________________________________________
#¿Para qué se usa esto? (El verdadero objetivo)
#Quien escribió ese código lo hizo para poder verificar si las variables son iterables o no. En Python, no todo se puede recorrer.

#Si completáramos ese código para hacer la prueba usando la función isinstance(), veríamos lo siguiente:

#from collections.abc import Iterable
#cadena = "Henry"
#numero = 9

# "Comprobemos la cadena"
print(isinstance(cadena, Iterable))  # Devuelve: True

# "Comprobemos el númer"o
print(isinstance(numero, Iterable))  # Devuelve: False
```
